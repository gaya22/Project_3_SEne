```python
# Importing necessary libraries
import pandas as pd

# Reading the CSV file
df = pd.read_csv('weather_data.csv')

# Displaying the first few rows of the dataframe
df.head()
```

```python
df.set_index('utc_timestamp', inplace=True)
df.index.name = 'date'

# Remove 'Z' at the end of the dates
df.index = pd.to_datetime(df.index.str[:-1])

df = df.loc['2015':]

df_monthly = df.resample('M').mean()
df_monthly.index = df_monthly.index.to_period('M').to_timestamp()

# Keep only the columns that contain '_temperature' after the first two characters
df_temperature = df_monthly[df.columns[df.columns.str[2:] == '_temperature']]

df_monthly_reindexed = df_temperature.reindex(columns=df.columns[df.columns.str[2:] == '_temperature'])



df_monthly_reindexed.head()


```

```python
# Dictionary mapping two-letter codes to European countries
country_mapping = {
    'AT': 'Austria',
    'BE': 'Belgium',
    'BG': 'Bulgaria',
    'CH': 'Switzerland',
    'CZ': 'Czech Republic',
    'DE': 'Germany',
    'DK': 'Denmark',
    'EE': 'Estonia',
    'ES': 'Spain',
    'FI': 'Finland',
    'FR': 'France',
    'GB': 'Great Britain',
    'GR': 'Greece',
    'HR': 'Croatia',
    'HU': 'Hungary',
    'IE': 'Ireland',
    'IT': 'Italy',
    'LT': 'Lithuania',
    'LU': 'Luxembourg',
    'LV': 'Latvia',
    'MT': 'Malta',
    'NL': 'Netherlands',
    'NO': 'Norway',  
    'PL': 'Poland',
    'PT': 'Portugal',
    'RO': 'Romania',
    'RU': 'Russia',
    'SE': 'Sweden',
    'SI': 'Slovenia',
    'SK': 'Slovakia',
    'UK': 'Great Britain',
    
    
}








# Iterate over columns and rename columns containing '_temperature'
for col in df_monthly_reindexed.columns:
    if '_temperature' in col:
        country_code = col[:2]
        country_name = country_mapping[country_code]
        new_col_name = country_name 
        df_monthly_reindexed.rename(columns={col: new_col_name}, inplace=True)

# Displaying the modified DataFrame
df_monthly_reindexed.head()

```

```python
df= pd.read_csv('estat_nrg_stk_gasm.tsv', sep='\t')
df = df.set_index(df.columns[0])



display(df)
```

```python
df = df.transpose()
df.index.name = 'date'







display(df)
```

```python
# Extracting and assigning the last two letters from column names
new_columns = [col.split(',')[-1][-2:] for col in df.columns]
df.columns = new_columns

df = df.replace(':', '0', regex=True)


display(df)
```
